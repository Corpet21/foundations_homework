```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors")
doc = BeautifulSoup(response.text)
```

```python
tables = doc.find_all("tr")
rows=[]

for table in tables:
    row = {}
    row["Rank"] = table.find("td", class_="number").get_text(strip=True)
    row["Title"] = table.find("a", class_= "bookTitle").get_text(strip=True)
    row["Author"] = table.find("div", class_="authorName__container").get_text(strip=True)
    row["Small_Text"] = table.find("span",class_="smallText uitext").get_text(strip=True)
    row["Rating"] = table.find("span", class_="greyText smallText uitext").get_text(strip=True)
    rows.append(row)
    print(row)
print(rows)
```

```python
df = pd.DataFrame(rows)
df
```

```python
columnsTitles = ['Rank', 'Title', 'Author', 'Rating', 'Small_Text']

df = df.reindex(columns=columnsTitles)
```

```python
df
```

```python
import re
```

```python
df['Votes'] = df['Small_Text'].str.extract('(.*and)', expand=True)
```

```python
df['Votes'] = df['Votes'].str.replace(",and","")
```

```python
df['Votes'] = df['Votes'].str.replace(",","")
```

```python
df['Votes'] = df['Votes'].str.replace("score:", "")
```

```python
regex_pat = re.compile(r'[(].*', flags=re.IGNORECASE)

df["Title"] = df["Title"].str.replace(regex_pat, '')
```

```python
regex_pat = re.compile(r'.*and', flags=re.IGNORECASE)

df["Small_Text"] = df["Small_Text"].str.replace(regex_pat, '')
```

```python
df["Small_Text"] = df["Small_Text"].str.replace("people voted", '')
```

```python
regex_pat = re.compile(r'avg.*', flags=re.IGNORECASE)

df["Rating"] = df["Rating"].str.replace(regex_pat, '')
```

```python
df["Author"] = df["Author"].str.replace("(Goodreads Author)", "")
```

```python
df.to_csv("goodreads.csv", index = False)
```
