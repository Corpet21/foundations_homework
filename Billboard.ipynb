```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.billboard.com/charts/hot-100')
doc = BeautifulSoup(response.text)
```

```python
rows = []

tables = doc.find_all("div", class_="chart-list-item__first-row chart-list-item__cursor-pointer")
for table in tables:
    row = {}
    row["rank"]= table.find(class_= 'chart-list-item__rank').text.strip()
    row['artist'] = table.find("div", class_ = "chart-list-item__artist").text.strip()
    row["track"] = table.find("span",class_= "chart-list-item__title-text").text.strip()
    rows.append(row)
print(rows)
        
```

```python
df = pd.DataFrame(rows)
df
```

```python
columnsTitles = ['rank', 'track', 'artist']

df = df.reindex(columns=columnsTitles)
```

```python
df.to_csv("billboard.csv", index = False)
```
