```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.epicurious.com/search/cucumbers")
doc = BeautifulSoup(response.text)
```

```python
rows = []

for page_num in range(1, 11):
    url = f"https://www.epicurious.com/search/cucumbers?page={page_num}"
    print(url)
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    recipes = doc.find_all("article")
    for recipe in recipes:
        row = {}
        try:
            try: 
                row['category'] = recipe.find(class_="tag").text.strip()
            except:
                pass
            try:
                row['title'] = recipe.find("h4", class_="hed").text.strip()
            except:
                pass
            try:
                row['summary']= recipe.find("p", class_="dek").text.strip()
            except:
                pass
            try: 
                row['rating']= recipe.find(class_="recipes-ratings-summary").text.strip()
            except: 
                pass
            try:
                row['would_make_again']=recipe.find(class_="make-again-percentage").text.strip()
            except:
                pass
            try:
                row['link']=recipe.find("a")["href"]
            except:
                pass
        except:
            pass
        print("----------")
        print(row)
        rows.append(row)

print("rows")
```

```python
df = pd.DataFrame(rows)
df.head()
```

```python
columnsTitles = ['category', 'title', 'summary', 'rating', 'would_make_again', "link"]

df = df.reindex(columns=columnsTitles)
```

```python
def scrape_page(row):
    url = "https://www.epicurious.com/" + row['link']
    print("Scraping", url)    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    page = {}
    page['ingredients'] = doc.find(class_="ingredients").text.strip()
    page['directions'] = doc.find(class_="instructions").text.strip()
    page['tags']= doc.find(class_="tags").text.strip()
    return pd.Series(page)
```

```python
scrape_page(df.loc[1])
```

```python
scrape_page(df.loc[2])
```

```python
scraped_df = df.apply(scrape_page, axis=1)
```

```python
df = df.merge(scraped_df, left_index= True, right_index= True)
```

```python
df.to_csv("Epicurius_Final", index=False)
```
